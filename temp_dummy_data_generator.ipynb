```python
import os
import pandas as pd
import numpy as np
import random
from collections import defaultdict
df = defaultdict(list)

pose_index = 0 
for trajectory_index in range(10):
    joint_angles = (np.ones(7,) * trajectory_index).tolist()
    effector_position = (np.ones(3,) * trajectory_index).tolist()
    df['joint_angles'].append(joint_angles)
    df['effector_position'].append(effector_position)
    df['trajectory_index'].append(trajectory_index)
    df['pose_index'].append(pose_index)
    pose_index += 1
    if random.random() < 0.5:
        df['joint_angles'].append(joint_angles)
        df['effector_position'].append(effector_position)
        df['trajectory_index'].append(trajectory_index)
        df['pose_index'].append(pose_index)
        pose_index += 1
        
df = pd.DataFrame.from_records(df)
if not os.path.exists('./data'): os.makedirs('./data')
```
